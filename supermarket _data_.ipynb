```python
import numpy
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
df=data.copy() 


```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['weekday'] = df['Date'].dt.day_name()
df['month'] = df['Date'].dt.month_name()

df.set_index('Date', inplace=True)
df.head()
```

```python

from sklearn.preprocessing import LabelEncoder

from sklearn.preprocessing import StandardScaler

from sklearn.model_selection import train_test_split

from sklearn.cluster import KMeans
```

```python
df1=df.copy() 
df1.head()
```

```python
df1.drop(columns=['Invoice ID','Time'],inplace=True)
```

```python
objList = df1.select_dtypes(include = "object").columns   
print (objList)        
```

```python
le = LabelEncoder()
for features in objList:
    df1[features] = le.fit_transform(df1[features])
df1.info()       
```

```python
df1.head(10) 
```

```python
scale = StandardScaler()
scaleddata = pd.DataFrame(scale.fit_transform(df1), columns=df1.columns)
```

```python
plt.figure(figsize=(15,8))
sn.heatmap(df1.corr(),annot=True,cmap=plt.cm.Reds).set_title('Correlation Heatmap.')
plt.show()
```

```python
model = KMeans(n_clusters=4, init="k-means++",random_state=0)
model.fit(df1)
```

```python
clusters = model.fit_predict(df1)
clusters
```

```python
plt.figure(figsize=(12,6))
sn.scatterplot(data=df1, x="Unit price", y="gross income", hue= clusters, palette = "bright")
centers = model.cluster_centers_
```

```python
clusters = pd.DataFrame(model.fit_predict(df1), columns=['clusters'])
richcustomers = data.loc[clusters['clusters'] == 1]
notrichcustomers = data.loc[clusters['clusters'] == 0]
```

```python
plt.figure(figsize=(15,5))
sn.set_palette(['red'])
sn.countplot(richcustomers['City'])
plt.show()
```

```python
plt.figure(figsize=(15,5))
sn.set_palette(['red'])
sn.countplot(richcustomers['Product line'])
plt.show()
```

```python
plt.figure(figsize=(15,5))
sn.set_palette(['red'])
sn.countplot(notrichcustomers['Product line'])
plt.show()
```

```python
plt.figure(figsize=(15,5))
sn.set_palette(['red'])
sn.distplot(richcustomers['Total'])
plt.show()
```

```python
plt.figure(figsize=(15,5))
sn.set_palette(['red'])
sn.distplot(notrichcustomers['Total'])
plt.show()
```

```python
richcustomers['Date'] = pd.to_datetime(richcustomers['Date'])
richcustomers['weekday'] = richcustomers['Date'].dt.day
plt.figure(figsize=(20,7))
sn.countplot(richcustomers['weekday'])
plt.show()
```

```python
notrichcustomers['Date'] = pd.to_datetime(notrichcustomers['Date'])
notrichcustomers['weekday'] = notrichcustomers['Date'].dt.day
plt.figure(figsize=(20,7))
sn.countplot(notrichcustomers['weekday'])
plt.show()
```

```python
#Clearly, what is clearly visible here is that the supermarket gets more visitors during particular days of the month.
#In the case of the wealthy customers the peak day of all three months is the 19th day of the respective month.
#in the case of the thrift shoppers, the peak day of each month is the 26th day of the month
```
